```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import seaborn as sns
```

```python
# pandas to open .csv files in raw folder
path = './raw/DISASTERS/'
import os
files = []
for file in os.listdir(path):
    print(file)
    files.append(file)
    
# read csv file

dataframes = []
for file in files:
    dataframes.append(pd.read_csv(path + '/' + file))
    


```

```python
dataframes[1].head()

disasters_df = pd.concat(dataframes, ignore_index=True)
```

```python
disasters_df
```

```python
disasters_df.isna().sum()
```

```python
# add Dis No if is nan
# Dis No is year-seq-ISO
sequence_format = disasters_df['Seq'].apply(lambda x: '{:0>4}'.format(x))
disasters_df['Dis No'] = disasters_df['Dis No'].fillna(disasters_df['Year'].astype(str) + '-' + sequence_format + '-' + disasters_df['ISO'].astype(str))
```

```python
# drop columns with more than 50% missing values
disasters_df_dropna = disasters_df.dropna(thresh=disasters_df.shape[0]*0.5,axis=1)
# drop duplicates
disasters_df_dropna = disasters_df_dropna.drop_duplicates(subset=['Dis No'], keep='first')
```

```python
disasters_df_dropna.isna().sum()
```

```python
# drop na values from Start Month column
disasters_df_dropna = disasters_df_dropna.dropna(subset=['Start Month'])
```

```python
disasters_df_dropna.isna().sum()
```

```python
# new datafram only with disasters from Chile

disasters_chile = disasters_df_dropna[disasters_df_dropna['Country'] == 'Chile']
```

```python
# translate disaster type to spanish
disasters_chile['Disaster Type'] = disasters_chile['Disaster Type'].replace({'Drought': 'Sequía', 'Earthquake': 'Terremoto', 'Epidemic': 'Epidemia', 'Extreme temperature': 'Temperatura extrema', 'Flood': 'Inundación', 'Insect infestation': 'Infestación de insectos', 'Landslide': 'Deslizamiento de tierra', 'Mass movement (dry)': 'Movimiento en masa (seco)', 'Storm': 'Tormenta', 'Volcanic activity': 'Actividad volcánica', 'Wildfire': 'Incendio forestal'})
# rename Disaster Type column to Tipo de desastre
disasters_chile = disasters_chile.rename(columns={'Disaster Type': 'Tipo de desastre'})
```

```python
# bar plot disasters per month

# One liner to create a stacked bar chart.
ax = sns.histplot(disasters_chile, x='Start Month', hue='Tipo de desastre', 
             multiple='stack', palette='tab20c', shrink=0.8, bins=12)
ax.set_ylabel('Cantidad de desastres')

ax.set_xlabel('Mes')
ax.set_xticks(np.arange(1, 13, 1))
ax.set_xticklabels(['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct','Nov', 'Dic'], rotation=45)
# set ticks middle of bars




ax.set_title('Desastres por mes en Chile')


# Fix the legend so it's not on top of the bars.
legend = ax.get_legend()
legend.set_bbox_to_anchor((1, 1))
```

```python
# bar plot disasters per month
disasters_chile_earthquake = disasters_chile[disasters_chile['Disaster Type'] == 'Earthquake']
# One liner to create a stacked bar chart.
ax = sns.histplot(disasters_chile_earthquake, x='Start Month', hue='Disaster Type', 
             multiple='stack', palette='tab20c', shrink=0.8, bins=12)
ax.set_ylabel('percentage')
# Fix the legend so it's not on top of the bars.
legend = ax.get_legend()
legend.set_bbox_to_anchor((1, 1))
```

```python
# Example data

# Define the order of the months
month_order = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct','Nov', 'Dic']

# Count the occurrences of each month and disaster type
counts = disasters_chile.groupby(['Start Month', 'Tipo de desastre']).size().unstack(fill_value=0)

# Transpose the counts DataFrame
counts = counts.T

# Set the colors for each disaster type
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan', 'black', 'yellow']
legend_labels = counts.index.tolist()

# Create the stacked bar chart
fig, ax = plt.subplots()

# Set the width of each bar
bar_width = 0.8

# Initialize the bottom positions for stacking
bottom = np.zeros(len(counts.columns))

# Plot each disaster type as a separate bar
for i, disaster_type in enumerate(counts.index):
    ax.bar(np.arange(len(counts.columns)), counts.loc[disaster_type], bottom=bottom, color=colors[i], width=bar_width, label=disaster_type)
    bottom += counts.loc[disaster_type]

# Set the x-axis tick positions and labels
ax.set_xticks(np.arange(len(counts.columns)))
ax.set_xticklabels(month_order)

# Set the y-axis label
ax.set_ylabel('Cantidad de desastres')

# Set the x-axis label
ax.set_xlabel('Mes')

# Set the title of the chart
ax.set_title('Desastres por mes en Chile')

# Create the legend
ax.legend(loc='upper right')

# Show the plot
plt.show()
```
